##Problem Statement

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
```

```python
dataframe = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Zomato Project/Zomato data .csv')
```

```python
dataframe.head(10)
```

```python
dataframe.isna().sum()
```

##Convert data type of columns Rating

```python
def handle_rating(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)

dataframe['rate'] = dataframe['rate'].apply(handle_rating)

dataframe.head()
```

```python
dataframe.info()
```

## Conclusion 1 - Majority of the restaurants falls under Dining category

```python
sns.countplot(x = 'listed_in(type)', data = dataframe, hue = 'listed_in(type)')
plt.xlabel('Type of Restaurant')
plt.title('Count of Restaurants by Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Conclusion 2 - Dining restaurants have more votes

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'Votes': grouped_data})
plt.plot(result,marker='o', c='green')
plt.xlabel('Type of restaurant')
plt.ylabel('Sum of Votes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the data and aggregate the votes by 'listed_in(type)'
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum().reset_index()

# Create a Seaborn barplot
sns.lineplot(x='listed_in(type)', y='votes', data=grouped_data)

# Add labels and title
plt.xlabel('Type of Restaurant')
plt.ylabel('Sum of Votes')
plt.title('Sum of Votes by Type of Restaurant')

# Show the plot
plt.show()
```

##Conclusion 3 - Majority of the restaurants received ratings from 3.5 to 3.7

```python
sns.histplot(x='rate',data=dataframe)
plt.show()
```

##Conclusion 4 - Most amount spent by couple on online orders is 500 & 600 Rs

```python
online_orders = dataframe[dataframe['online_order'] == 'Yes']
```

```python
most_frequent_spending_online = online_orders['approx_cost(for two people)'].mode()
print(f"The most couples are spending on online orders is: {', '.join(map(str, most_frequent_spending_online))} Rs")
```

```python
sns.countplot(x=online_orders['approx_cost(for two people)'])
plt.title('Distribution of Spending for Online Orders')
plt.xlabel('Spending Amount')
plt.ylabel('Number of Couples')
plt.show()
```

##Conclusion 5 - Online mode received maximum ratings in comparision to offline ratings

```python
sns.boxplot(hue ='online_order',y='rate',data=dataframe, legend=True)
plt.xlabel('online_order')
plt.show()
```

##Conclusion 6 - Dining Type has received more orders in offline mode since it's mainly based on live expereience where as for online cafes received more orders

```python
# order_type_heatmap = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size')
order_counts = dataframe.groupby(['listed_in(type)', 'online_order']).size().unstack(fill_value=0)

# Plot heatmap
plt.figure(figsize=(10, 6))  # Set figure size
sns.heatmap(order_counts, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Count'})

# Add title and labels
plt.title('Heatmap of Order Types by Category', fontsize=16)
plt.xlabel('Online Order', fontsize=12)
plt.ylabel('Category (Type)', fontsize=12)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
